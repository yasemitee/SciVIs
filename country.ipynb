```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
#Lettura dei dati e creazione dei dataframe per i giocatori, i tornei e i match ATP dal 1968 al 2024 
atp_matches = pd.read_csv('Data/atpdata/atp_matches.csv')
atp_players = pd.read_csv('Data/atpdata/atp_players.csv')
atp_rankings = pd.read_csv('Data/atpdata/atp_rankings.csv', low_memory=False);

atp_matches['tourney_date'] = pd.to_datetime(atp_matches.tourney_date, format='%Y%m%d')



#Pallet di colori per i grafici
colors = ['#11565c', '#007660', '#459146', '#9aa414', '#ffa600' ]
```

```python
atp_matches_country = atp_matches[['winner_ioc', 'tourney_level', 'round']]
atp_matches_country = atp_matches_country[(atp_matches_country['tourney_level'] == 'G')]

atp_matches_country.count()

plt.figure(figsize=(150, 8))
plt.subplot(1,9,1)
plt1 = atp_matches_country['winner_ioc'].value_counts()[:-30].plot(kind='bar', color=colors)
plt.title('TOTAL GRAND SLAM MATCH WINS')
plt1.set(xlabel = 'Country', ylabel='Wins')
```

#### Giocatori migliori per top 5 paesi più vincenti

```python

usa = atp_matches[['winner_ioc', 'tourney_level', 'winner_name', 'tourney_date']]


usa = usa[(usa['winner_ioc'] == 'USA') & (usa['tourney_level'] == 'G')]

count_usa = usa['winner_name'].value_counts()

# grafico a barre di count_usa dei migliori 5 giocatori
plt.figure(figsize=(100, 7))
plt.subplot(1,9,1)
plt1 = count_usa[:5].plot(kind='bar', color=colors)
plt.title('GRAND SLAM WINS - USA')
plt1.set(xlabel = 'Player', ylabel='Wins')

for i in range(len(count_usa[:5])):
    plt.text(i, count_usa[i], count_usa[i], ha = 'center', va = 'bottom')




```
